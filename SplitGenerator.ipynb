### Split each song into ten-second sub recording wave files, and label each split with the name of the ragam of the song from which the split is cut.

```python
#audio, sr = librosa.load(file_name)
def splitSong(sample_rate, data1, ragam):
# Get number of samples for 2 seconds; replace 2 by any number

  ragam = ragam.lower()

  buffer = 10 * sample_rate

  samples_total = len(data1) 
  samples_wrote = 0
  index = ragamDict[ragam] + 1
  numSplits = 0
  while samples_wrote < samples_total:

      #check if the buffer is not exceeding total samples 
      if buffer > (samples_total - samples_wrote):
          buffer = samples_total - samples_wrote

      block = data1[samples_wrote : (samples_wrote + buffer)]
      
      pathName = "/content/drive/MyDrive/N240Project/RagamSplits3/"
      out_filename = pathName + ragam.lower() + "/" + ragam + "_" + "split_" + str(index)

      # Write 10 second segment
      sf.write(out_filename, block, sample_rate, format = 'WAV')
      index += 1
      numSplits += 1
      samples_wrote += buffer

  ragamDict[ragam] += numSplits
```
